```python
def sort_orders():  
    print('Make sure the file is an excel document without macros.')
    file_name = input('enter file name: ') + '.xlsx'

    import pandas as pd
    import datetime as date 

    try:
        df = pd.read_excel(file_name) 
        df.head()  
        print('Your orders are being sorted. Please wait...')
    except: 
        print('''The File you entered cannot be found. Reasons for this may include:
                 1. forgot capital letters 
                 2. file is not located in this directory 
                 3. file is not an excel workbook 
                 4. file is saved as a macro workbook''')
        
    today = date.date.today() 
    d = today.strftime("%m-%d-%Y")
    mrp_c = ['PBR', 'PHY', 'PPH', 'PAI', 'PST', 'P14', 'PHV', 'PPK', 'PPT',
       'PTI', 'PHE', 'PSE', 'PLE', 'PS3', 'PGP', 'PPP', 'PAA', 'PUB',
       'PGU', 'PHC', 'PCP', 'PMA', 'PSK', 'PKG', 'PBM', 'PPL', 'PCL',
       'PXT', 'PBT','PS2'] 
    ar = df['MRP controller'].unique() 
    for i in ar: 
        with pd.ExcelWriter("UNKOWN ORDERS {}.xlsx".format(today)) as w:
            if i not in mrp_c: 
                df.loc[df['MRP controller'] == i].to_excel(w, sheet_name=str(i))  
            else: 
                continue
        
    #SEP's
    sep_main = df.loc[df['MRP controller'] == 'PSE']
    sep_ex = df.loc[df['MRP controller'] == 'PS3'] 
    sep_ng = df.loc[df['MRP controller'] == 'PS2'] 
    
    #LEP HYDRA AIR HELIX
    lep = df.loc[df['MRP controller'] == 'PLE']
    hydra = df.loc[df['MRP controller'] == 'PHY']   
    air = df.loc[df['MRP controller'] == 'PAI'] 
    helix = df.loc[df['MRP controller'] == 'PPH'] 
    
    #HVLP PP AA 
    hvlp = df.loc[df['MRP controller'] == 'PHV']  
    pp = df.loc[df['MRP controller'] == 'PPP'] 
    aa = df.loc[df['MRP controller'] == 'PAA']
    
    #SKN STM GUN  
    stm = df.loc[df['MRP controller'] == 'PST'] 
    gun = df.loc[df['MRP controller'] == 'PGU'] 
    skn = df.loc[df['MRP controller'] == 'PSK']
    #get last one
    
    #PT PL GPX 
    gpx = df.loc[df['MRP controller'] == 'PGP']
    pt = df.loc[df['MRP controller'] == 'PPT']
    pl = df.loc[df['MRP controller'] == 'PPL']
    
    #TIP 
    tip = df.loc[(df['MRP controller'] == 'PPK') | 
                 (df['MRP controller'] == 'PTI') | 
                 (df['MRP controller'] == 'PHE') | 
                 (df['MRP controller'] == 'PKG')] 
    
    #Sub and others 
    #SUB 
    sub = df.loc[df['MRP controller'] == 'PUB']
    
    #MAC 
    mac = df.loc[df['MRP controller'] == 'PMA']
    
    #BRZ 
    brz = df.loc[df['MRP controller'] == 'PBR']
    
    #CP 
    cp = df.loc[df['MRP controller'] == 'PCP'] 
    
    #other
    P14 = df.loc[df['MRP controller'] == 'P14'] 
    PHC = df.loc[df['MRP controller'] == 'PHC']
    PSK = df.loc[df['MRP controller'] == 'PSK']
    PKG = df.loc[df['MRP controller'] == 'PKG']
    PBM = df.loc[df['MRP controller'] == 'PBM']
    PCL = df.loc[df['MRP controller'] == 'PCL']
    PXT = df.loc[df['MRP controller'] == 'PXT']
    PBT = df.loc[df['MRP controller'] == 'PBT']
    
    try: 
        with pd.ExcelWriter("SEP {}.xlsx".format(today)) as writer1:
            sep_main.to_excel(writer1, sheet_name="SEP Main")  
            sep_ex.to_excel(writer1, sheet_name="SEP Express") 
            sep_ng.to_excel(writer1, sheet_name='SEP Next Gen')
 
        with pd.ExcelWriter("LEP Hydra Air Orders {}.xlsx".format(today), datetime_format='str') as writer2:
                lep.to_excel(writer2, sheet_name='LEP') 
                hydra.to_excel(writer2, sheet_name='Hydra') 
                air.to_excel(writer2, sheet_name='Air') 
                helix.to_excel(writer2, sheet_name='Helix')
 
        with pd.ExcelWriter("LEP Hydra Air Orders {}.xlsx".format(today)) as writer3:
                hvlp.to_excel(writer3, sheet_name='HVLP') 
                pp.to_excel(writer3, sheet_name='PP') 
                aa.to_excel(writer3, sheet_name='AA')  
        
        with pd.ExcelWriter('SKN STM Gun {}.xlsx'.format(today)) as writer4: 
            stm.to_excel(writer4, sheet_name='Steam') 
            gun.to_excel(writer4, sheet_name='Gun')
            skn.to_excel(writer4, sheet_name='skn' ) 
            
        with pd.ExcelWriter('PT PL GPX {}.xlsx'.format(today)) as writer5: 
            gpx.to_excel(writer5, sheet_name='Gpx') 
            pt.to_excel(writer5, sheet_name='PT') 
            pl.to_excel(writer5, sheet_name='PL') 
        
        with pd.ExcelWriter('PT PL GPX {}.xlsx'.format(today)) as writer6:
            tip.to_excel(writer6, sheet_name='tip') 
            
        with pd.ExcelWriter('SUB and others {}.xlsx'.format(today)) as writer7:
            sub.to_excel(writer7, sheet_name='SUB')
            mac.to_excel(writer7, sheet_name='mac')
            brz.to_excel(writer7, sheet_name='brz')
            cp.to_excel(writer7, sheet_name='cp')
            P14.to_excel(writer7, sheet_name='P14') 
            PHC.to_excel(writer7, sheet_name='PHC')
            PSK.to_excel(writer7, sheet_name='PSK')
            PKG.to_excel(writer7, sheet_name='PKG')
            PBM.to_excel(writer7, sheet_name='PBM')
            PCL.to_excel(writer7, sheet_name='PCL')
            PXT.to_excel(writer7, sheet_name='PXT')
            PBT.to_excel(writer7, sheet_name='PBT') 
            
        return print('Orders have been successfully sorted. Have a good day.')
            
    except: 
        raise print('''There has been an error while executing the program. Please double check your file name 
                       and the code before trying again.''') 
```

```python
sort_orders()
```
